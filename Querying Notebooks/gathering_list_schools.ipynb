```python
import pandas as pd

# ── 1. Load the public schools file ─────────────────────────────────────────
df_schools = pd.read_excel('../raw_data/pubschls.xlsx', dtype=str, header=5)

# ── 2. Keep only active schools ──────────────────────────────────────────────
df_schools = df_schools[df_schools['StatusType'] == 'Active'].copy()

# ── 3. Filter to schools that offer up to grade 12 (true high schools) ───────
# GSoffered must end in 12 (e.g. 9-12, 10-12, 7-12) but NOT start below 6
# This excludes K-12 and elementary/middle combos while keeping pure high schools
#high_school_patterns = ['9-12', '10-12', '11-12', '9-11', '10-11', '9-10', '10-Adult', '9-Adult']
high_school_patterns = ['9-12']
df_hs = df_schools[df_schools['GSserved'].isin(high_school_patterns)].copy()

# ── 4. Extract the 7-digit School Code from CDSCode ──────────────────────────
# CDSCode format is: 2-digit County + 5-digit District + 7-digit School
df_hs['County Code'] = df_hs['CDSCode'].str[:2]
df_hs['District Code'] = df_hs['CDSCode'].str[2:7]
df_hs['School Code'] = df_hs['CDSCode'].str[7:]
```

```python
import json

school_codes_list = df_hs['School Code'].to_list()

with open("school_codes_list.json", "w") as f:
    json.dump(school_codes_list, f)
```
