```python
import json

with open("school_codes_list.json", "r") as f:
    school_codes_list = json.load(f)
```

```python
import pandas as pd
import requests
from io import StringIO

url = "https://www3.cde.ca.gov/demo-downloads/homeless/hse2425.txt"

r = requests.get(url)
r.raise_for_status()

text = r.content.decode("cp1252", errors="replace")

df = pd.read_csv(StringIO(text), sep="\t")
df.columns = df.columns.str.strip()
```

```python
homeless_df = df[df["Aggregate Level"] == "S"].copy()
homeless_df = homeless_df[homeless_df["Reporting Category"] == "TA"].copy()

homeless_df["School Code"] = homeless_df["School Code"].astype(int).astype(str).str.zfill(7)

homeless_df = homeless_df[homeless_df["School Code"].isin(school_codes_list)].copy()

homeless_df['Homeless Student Enrollment (Rate)'] = homeless_df['Homeless Student Enrollment'] / homeless_df['Cumulative Enrollment']
```

```python
homeless_df = homeless_df[['School Code', 'Homeless Student Enrollment (Rate)']]
```
