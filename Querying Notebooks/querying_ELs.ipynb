```python
import json

with open("school_codes_list.json", "r") as f:
    school_codes_list = json.load(f)
```

```python
import pandas as pd

df = pd.read_csv(
    "../raw_data/ELs.txt",
    sep="\t",
    dtype={"CDS": str}
)

df.columns = df.columns.str.strip()

df.head()
```

```python
df["CountyCode"] = df["CDS"].str[:2]
df["DistrictCode"] = df["CDS"].str[2:7]
df["SchoolCode"] = df["CDS"].str[7:]
```

```python
df = df[df["SchoolCode"].isin(school_codes_list)].copy()
```

```python
school_el_totals = (
    df.groupby("SchoolCode", as_index=False)["TOTAL_EL"]
    .sum()
)
```

```python
school_el_totals.to_csv('../cleaned_data/EL_data.csv')
```
