```python
import json

with open("school_codes_list.json", "r") as f:
    school_codes_list = json.load(f)
```

```python
import pandas as pd

url = "https://www3.cde.ca.gov/demo-downloads/acgr/acgr25.txt"

df = pd.read_csv(url, sep="\t")

df.columns = df.columns.str.strip()
```

```python
df = df[df["AggregateLevel"] == "S"].copy()

df = df[df["ReportingCategory"] == "TA"].copy()

df["SchoolCode"] = df["SchoolCode"].astype(int).astype(str).str.zfill(7)

df = df[df["SchoolCode"].isin(school_codes_list)].copy()
```

```python
cols = [
    "SchoolCode",
    "Met UC/CSU Grad Req's (Rate)",
    "Dropout (Rate)"
]

df = df[cols]
```
