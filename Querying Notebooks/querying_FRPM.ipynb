```python
import json

with open("school_codes_list.json", "r") as f:
    school_codes_list = json.load(f)
```

```python
import pandas as pd

df = pd.read_excel(
    "../raw_data/frpm2425.xlsx",
    sheet_name="FRPM School-Level Data",
    skiprows=1
)

df.columns = df.columns.str.strip()
df["School Code"] = df["School Code"].astype(str).str.zfill(7)
```

```python
df = df[df["School Code"].isin(school_codes_list)].copy()
```

```python
df = df[['School Code', 'Percent (%) \nEligible FRPM \n(K-12)', 'School Type', 'Educational \nOption Type']]
```

```python
df = df.rename(columns={'Percent (%) \nEligible FRPM \n(K-12)': 'Percent Eligible FRPM', 'Educational \nOption Type': 'Educational Option Type'})
```
