```python
import pandas as pd
from functools import reduce
import numpy as np
```

```python
df1 = pd.read_csv('../cleaned_data/absenteeism_data.csv').drop(columns=['Unnamed: 0'])
if "School Code" in df1.columns:
    df1.rename(columns={"School Code": "SchoolCode"}, inplace=True)
```

```python
df_list = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11]

merged_df = reduce(lambda left, right: pd.merge(left, right, on="SchoolCode"), df_list)
```

```python
merged_df = merged_df.drop(columns=['County Code', 'District Code'])
```

```python
merged_df = merged_df[merged_df['School Type'] == 'High Schools (Public)']
```

```python
merged_df = merged_df.replace("*", np.nan)
```
