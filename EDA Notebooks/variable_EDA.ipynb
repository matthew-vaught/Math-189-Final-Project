```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#We only look at Public California Schools
schools = pd.read_csv('../cleaned_data/final_merged_dataset.csv').drop(columns=['Unnamed: 0'])
schools = schools[schools["School Type"] == "High Schools (Public)"]
schools.head(5)
```

```python
schools.columns
```

## Single Variable Analysis

### Distribution of College Going Rate

```python
college_going_rate = schools["College Going Rate - Total (12 Months)"]

sns.histplot(college_going_rate, bins=20, kde=True, color="darkblue")

plt.xlabel("College Going Rate (%)")
plt.title("Distribution of College Going Rates")
plt.show()

# summary statistics
schools["College Going Rate - Total (12 Months)"].describe()
```

### Distribution of Proportion of Students Experiencing Homelessness

```python
homelessness_rate = schools["Homeless Student Enrollment (Rate)"]

sns.histplot(homelessness_rate, bins=20, kde=True, color="darkgoldenrod")

plt.xlabel("Homelessness Proportion")
plt.title("Distribution of Proportion of Homeless Students")
plt.show()

# summary statistics
schools["Homeless Student Enrollment (Rate)"].describe()
```

### Distribution of Explusions

```python
expulsions = schools["Expulsion Rate"]

sns.histplot(expulsions, bins=20, kde=True, color="darkorange")

plt.xlabel("Rate of Expulsions")
plt.title("Distribution of Rate of Expulsions")
plt.show()

# summary statistics
expulsions.describe()
```

### Distribution of Suspensions

```python
suspensions = schools["Suspension Rate"]

sns.histplot(suspensions, bins=20, kde=True, color="purple")

plt.xlabel("Rate of Suspensions")
plt.title("Distribution of Rate of Suspensions")
plt.show()

# summary statistics
suspensions.describe()
```

### Distribution of Student-Teacher Ratios

```python
student_teacher_ratio = schools["STU_TCH_RATIO"]

sns.histplot(student_teacher_ratio, bins=20, kde=True, color="darkgreen")

plt.xlabel("Student Teacher Ratio")
plt.title("Distribution of Student to Teacher Ratio")
plt.show()

# summary statistics
schools["STU_TCH_RATIO"].describe()
```

### Distribution of Proportion of Students eligible for FRPM (Free/Reduced Price Meals)

```python
frpm_prop = schools["Percent Eligible FRPM"]

sns.histplot(frpm_prop, bins=20, kde=True, color="darkred")

plt.xlabel("FRPM Proportion")
plt.title("Distribution of Proportion of Students Eligible for FRPM")
plt.show()

# summary statistics
schools["Percent Eligible FRPM"].describe()
```

### Distribution of Chronic Absentees

```python
absentees = schools["ChronicAbsenteeismRate"]

sns.histplot(absentees, bins=20, kde=True, color="darkgray")

plt.xlabel("Number of Absentees")
plt.title("Distribution of Chronic Absentees")
plt.show()

# summary statistics
schools["ChronicAbsenteeismRate"].describe()
```

### Distribution of Percentage of Students Meeting or Exceeding State Standards

```python
standards = schools["Percentage Standard Met and Above"]

sns.histplot(standards, bins=20, kde=True, color="darkviolet")

plt.xlabel("Students Met or Above Standard (%)")
plt.title("Distribution of Percentage of Students Meeting or Exceeding State Standards")
plt.show()

# summary statistics
schools["Percentage Standard Met and Above"].describe()
```

### Distribution of Percentage of Students Enrolled for the Entire Academic Year

```python
stability_rate = schools["Stability Rate (percent)"]

sns.histplot(stability_rate, bins=20, kde=True, color="darkblue")

plt.xlabel("Percentage of Students Enrolled for the Entire Academic Year")
plt.title("Distribution of Stability Rate Percent")
plt.show()

# summary statistics
schools["Stability Rate (percent)"].describe()
```

### Boxplots for All Variables

```python
vars_to_plot = [
    "College Going Rate - Total (12 Months)",
    "Homeless Student Enrollment (Rate)",
    "Expulsion Rate",
    "Suspension Rate",
    "STU_TCH_RATIO",
    "Percent Eligible FRPM",
    "ChronicAbsenteeismRate",
    "Percentage Standard Met and Above",
    "Stability Rate (percent)"
]

df = schools[vars_to_plot]

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(vars_to_plot):
    sns.boxplot(x=df[col], ax=axes[i])
    axes[i].set_title(f"Boxplot of {col}")
    axes[i].set_xlabel(col)

plt.tight_layout()
plt.show()
```

## Scatterplots between College Going Rates and Features of Interest

### Scatterplot between College Going Rate and Homeless Rate

```python
y = "College Going Rate - Total (12 Months)"
x = "Homeless Student Enrollment (Rate)"

corr = schools[[x, y]].corr().iloc[0,1]
print(f"Correlation: {corr:.3f}")

sns.scatterplot(data=schools, x=x, y=y, color="darkgoldenrod")
plt.title(f"{y} vs {x} (corr = {corr:.2f})")
plt.show()
```
